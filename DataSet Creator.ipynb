```python
import pandas as pd
import os
import shutil
```

```python
# Create the data for positive samples

FILE_PATH = "C:\\Users\\Freaky\\Desktop\\chestxray\\metadata.csv"
IMAGE_PATH = "C:\\Users\\Freaky\\Desktop\\chestxray\\images"
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head()
```

```python
# Target directory to store images
TARGET_DIR = "C:\\Users\\Freaky\\Desktop\\Dataset\\Covid"
if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
```

```python
# Place images of covid positive samples in the folder that was created
cnt = 0 

for (i,row) in df.iterrows():
    if row["finding"]=="COVID-19" and row["view"]=="PA":
        filename = row["filename"]
        image_path = os.path.join(IMAGE_PATH,filename)
        image_copy_path = os.path.join(TARGET_DIR,filename)
        shutil.copy2(image_path,image_copy_path)
        print("Moving images",cnt)
         cnt += 1
        
print(cnt)
```

```python
# Sampling of images from Kaggle

import random
KAGGLE_FILE_PATH = "C:\\Users\\Freaky\\Desktop\\chest_xray_kaggle\\chest_xray\\train\\NORMAL"
TARGET_NORMAL_DIR = "C:\\Users\\Freaky\\Desktop\\Dataset\\Normal"
```

```python
image_names = os.listdir(KAGGLE_FILE_PATH)
```

```python
#Shuffle all 6000 images
random.shuffle(image_names)
```

```python
#Take 206 random images from 6000
for i in range(206):
    image_name = image_names[i]
    image_path = os.path.join(KAGGLE_FILE_PATH,image_name)
    target_path = os.path.join(TARGET_NORMAL_DIR,image_name)
    shutil.copy2(image_path,target_path)
    print("copying image ",i)
```
